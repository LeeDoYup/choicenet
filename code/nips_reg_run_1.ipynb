### Initiate Workers

```python
nWorker = 4
maxGPU  = 2
WORKERS = ['']*nWorker 
for i in range(nWorker):
    WORKERS[i] = worker_class(_idx=i,_maxProcessID=nWorker,_maxGPU=maxGPU
                              ,_name='Worker_%02d'%(i),_VERBOSE=False)
    print ("[%d/%d] worker initiated."%(i,nWorker))
```

### Start Workers

```python
for i in range(nWorker):
    WORKERS[i].start(); # Start process
```
