#### Outlier 60% / measurement noise: 1e-6 / tauInv=1e-2

```python
kmixList = [2,3,4,5,8,10,15,20]
rmseList = [0.18,0.10,0.07,0.05,0.06,0.06,0.06,0.06]
plt.figure(figsize=(6,5))
plt.plot(kmixList,rmseList,'o-')
plt.xlabel('K',fontsize=20)
plt.ylabel('RMSE',fontsize=20)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.show()
```
